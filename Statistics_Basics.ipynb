**Question 1: What is the difference between descriptive statistics and inferential
statistics? Explain with examples.**

    Descriptive statistics summarize and describe the main features of a dataset (for example: mean, median, mode, variance, standard deviation, histograms, boxplots). They provide a compact description of the observed data only. Inferential statistics use a sample of data to make conclusions or predictions about a larger population; they quantify uncertainty using techniques like confidence intervals, hypothesis tests, regression, and p-values.

**Question 2: What is sampling in statistics?**

    Explain random vs stratified sampling. Sampling is selecting a subset of observations from a population to estimate characteristics of the whole population. Random (simple random) sampling means each individual has an equal chance to be selected; it reduces selection bias and supports standard inferential formulas. Stratified sampling first divides the population into homogeneous subgroups (strata) and then samples from each stratum (often proportionally). Stratified sampling increases precision when strata differ on the measured variable (ensures representation from each subgroup).

**Question 3: Define mean, median, and mode. Why are they important?**

    Mean (arithmetic mean) is the sum of all values divided by the count. It measures the average.

    Median is the middle value when data are sorted (or average of two middle values for even count). It is robust to outliers.

    Mode is the most frequently occurring value(s) in the dataset. Useful for categorical data or multimodal distributions.
    These measures are important because they describe the central tendency of a dataset and help summarize and compare datasets.

**Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?**

    Skewness measures distribution asymmetry. Positive skew (right skew) means the right tail is longer — more extreme large values exist. Typically mean > median in positive skew.

    Kurtosis measures tail weight/peakedness relative to normal distribution. High kurtosis (leptokurtic) indicates heavier tails and more extreme outliers; low kurtosis (platykurtic) indicates lighter tails.
    Positive skew implies that the bulk of the data is on the left and there are relatively few large values (outliers) to the right.

**Question 5: Implement a Python program to compute mean, median, and mode of
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]**

```python
import statistics as stats
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]
mean_val = stats.mean(numbers)
median_val = stats.median(numbers)
mode_val = stats.multimode(numbers)  # supports multimodal

print("Mean =", mean_val)
print("Median =", median_val)
print("Mode(s) =", mode_val)
```


**Question 6: Compute covariance and correlation between
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]**

```python
import numpy as np
list_x = [10,20,30,40,50]
list_y = [15,25,35,45,60]
cov_xy = np.cov(list_x, list_y, ddof=1)[0,1]     # sample covariance
corr_xy = np.corrcoef(list_x, list_y)[0,1]      # Pearson correlation
print("Covariance =", cov_xy)
print("Correlation =", corr_xy)
```

**Question 8: Relationship between advertising_spend and daily_sales
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]**

    Explanation and code:
    Use covariance to check direction (positive/negative) and Pearson correlation to measure strength (-1 to 1). High positive correlation suggests advertising spend and daily sales increase together and more spending is associated with higher sales.

```python
import numpy as np
advertising_spend = [200,250,300,400,500]
daily_sales = [2200,2450,2750,3200,4000]
cov = np.cov(advertising_spend, daily_sales, ddof=1)[0,1]
corr = np.corrcoef(advertising_spend, daily_sales)[0,1]
print("Covariance =", cov)
print("Correlation =", corr)
```

**Question 9: Customer survey histogram
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]**

```python
import numpy as np
survey_scores = [7,8,5,9,6,7,8,9,10,4,7,6,9,8,7]
mean_survey = np.mean(survey_scores)
std_survey = np.std(survey_scores, ddof=1)
print("Mean =", mean_survey)
print("Sample Std Dev =", std_survey)

# to plot:
import matplotlib.pyplot as plt
plt.hist(survey_scores, bins=range(4,12), align='left', edgecolor='black'); plt.show()
```
